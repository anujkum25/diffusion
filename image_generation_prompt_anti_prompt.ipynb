```python
from diffusers import StableDiffusionPipeline,StableDiffusionXLPipeline
import torch
```

```python
#pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1", torch_dtype=torch.float16).to("cuda")

pipe = StableDiffusionPipeline.from_pretrained("SG161222/Realistic_Vision_V6.0_B1_noVAE", torch_dtype=torch.float16).to("cuda")

#pipe = StableDiffusionXLPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0",torch_dtype=torch.float16).to("cuda")

```

```python
import pandas as pd

dataset1= pd.read_csv("/content/dataset1.csv")
dataset1.head()
```

```python
for i in range(3):
  prompt = dataset1["prompt"][i]
  anti_prompt = dataset1["test prompt"][i]

  for j in range(3):
    image_prompt = pipe(prompt).images[0]
    image_anti_prompt = pipe(anti_prompt).images[0]

    image_prompt.save(f"/content/sample_data/Output/{prompt}_{i}_{j}th_iamge.png")
    image_anti_prompt.save(f"/content/sample_data/Output/{anti_prompt}_{i}_{j}th_iamge.png")


```
