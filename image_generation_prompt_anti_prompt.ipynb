#download real images from web for same prompt and antiprompt

```python
#!pip install requests beautifulsoup4 pillow
import requests
from bs4 import BeautifulSoup
from PIL import Image
from io import BytesIO
import os
```

```python
def download_images(url, output_folder):
    response = requests.get(url)
    image=Image.open(BytesIO(response.content))
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    image_name= (os.path.join(output_folder, url.split('/')[-1]))
    image.save(image_name)
    return image_name


def extract_image_urls(webpage_url):
    response = requests.get(webpage_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    img_tags = soup.find_all('img')
    image_urls = [img['src'] for img in img_tags if 'src' in img.attrs]
    return image_urls

```

```python
webpage_url = 'https://www.thoughtco.com/who-are-the-brahmins-195316'

img_urls = extract_image_urls(webpage_url)

for img_url in img_urls:
  print(img_url)
  download_images(img_url, '/content/drive/MyDrive/diffusion_real_downloaded')
```
